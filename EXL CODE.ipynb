```python
import pandas
URL = 'train.xlsx'
df_Demo = pandas.read_excel(URL,sheetname=0)
df_cd = pandas.read_excel(URL,sheetname=1)
df_reld = pandas.read_excel(URL,sheetname=2)
df_defd = pandas.read_excel(URL,sheetname=3)
df_usep = pandas.read_excel(URL,sheetname=4)
```

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


df_dd = df_Demo[:]
df_dd['Churn Date'].fillna(0, inplace=True)
X = df_dd['Churn Date'].values != 0
X = X +[0]
df_dd['Categ'] = X


query = []
from collections import Counter
Dict = Counter(df_cd['Account Number'].values)
for i in df_dd['Account Number'].values:
    query.append(Dict[i])
df_dd['Query_freq'] = query


mon = []
for i in range(len(df_reld['Realization Date'].values)):
        mon.append(pd.Timestamp(df_reld['Realization Date'].values[i].astype(object)).to_pydatetime().month)
df_reld['mon'] = mon
x = df_reld.groupby('Account Number')['mon'].apply(list)
Realization_monnos = []
for i in df_dd['Account Number'].values:
    if i in df_reld['Account Number'].values:
        Realization_monnos.append(len(np.unique(x[i])))
    else:
        Realization_monnos.append('None')
df_dd['Realization_monnos'] = Realization_monnos



Default_number = []
from collections import Counter
Dict = Counter(df_defd['Account Number'].values)
for i in df_dd['Account Number'].values:
    Default_number.append(Dict[i])
df_dd['Default_number'] = Default_number



Commence_year = []
Commence_month = []
Commence_day = []
for i in range(len(df_dd)):
    if str(df_dd['Commence Date'].values[i]) == 'NaT':
                            Commence_year.append(np.nan)
                            Commence_month.append(np.nan)
                            Commence_day.append(np.nan)
    else:

                            Commence_year.append(pd.Timestamp(df_dd['Commence Date'].values[i].astype(object)).to_pydatetime().year)
                            Commence_month.append(pd.Timestamp(df_dd['Commence Date'].values[i].astype(object)).to_pydatetime().month)
                            Commence_day.append(pd.Timestamp(df_dd['Commence Date'].values[i].astype(object)).to_pydatetime().day)
                            
                        
df_dd['Commence_year'] = Commence_year
df_dd['Commence_month'] = Commence_month
df_dd['Commence_day'] = Commence_day



Commence_year = []
Commence_month = []
Commence_day = []
for i in range(len(df_dd)):
    if str(df_dd['Equipment Warranty Expiry Date'].values[i]) == 'NaT':
                            Commence_year.append(np.nan)
                            Commence_month.append(np.nan)
                            Commence_day.append(np.nan)
    else:

                            Commence_year.append(pd.Timestamp(df_dd['Commence Date'].values[i].astype(object)).to_pydatetime().year)
                            Commence_month.append(pd.Timestamp(df_dd['Commence Date'].values[i].astype(object)).to_pydatetime().month)
                            Commence_day.append(pd.Timestamp(df_dd['Commence Date'].values[i].astype(object)).to_pydatetime().day)
                            
                        
df_dd['Warrenty_year'] = Commence_year
df_dd['Warrenty_month'] = Commence_month
df_dd['Warrenty_day'] = Commence_day

DICT_1 = df_reld.set_index('Account Number').T.to_dict('list')
Cardtype = []
schemecost = []
for i in df_dd['Account Number'].values:
    if i in df_reld['Account Number'].values:
        Cardtype.append(DICT_1[i][2])
        schemecost.append(DICT_1[i][1])
    else:
        Cardtype.append('None')
        schemecost.append('None')        
df_dd['Cardtype'] = Cardtype
df_dd['schemecost'] = schemecost

DICT_2 = df_defd.set_index('Account Number').T.to_dict('list')
Cardtype = []
schemecost = []
for i in df_dd['Account Number'].values:
    if i in df_defd['Account Number'].values:
        Cardtype.append(DICT_2[i][0])
        schemecost.append(DICT_2[i][1])
    else:
        Cardtype.append('None')
        schemecost.append('None')        
df_dd['Closetdefaultdate'] = Cardtype
df_dd['defaultsum'] = schemecost

x = df_usep.groupby(['Account Number '], as_index=False).sum()
DICT_3 = x.set_index('Account Number ').T.to_dict('list')
Cardtype = []
schemecost = []
daraGB = []
YYYYMM = []

for i in df_dd['Account Number'].values:
    if i in df_usep['Account Number '].values:
        Cardtype.append(DICT_3[i][0])
        schemecost.append(DICT_3[i][1])
        daraGB.append(DICT_3[i][2])
        YYYYMM.append(DICT_3[i][3])
    else:
        Cardtype.append('None')
        schemecost.append('None')
        daraGB.append('None')        
        YYYYMM.append('None')        
df_dd['Hours'] = Cardtype
df_dd['Posthours'] = schemecost
df_dd['daraGB'] = daraGB
df_dd['YYYYMM'] = YYYYMM
```
